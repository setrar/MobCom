```julia
Œ∏‚ÇÄ  = 0    # Phase shift in radians : radians to degrees -> Œ∏ * 180/œÄ

# Generate the cosine wave w‚ÇÄ
w‚ÇÄ = A * cos.(2œÄ * ùëì‚Çç .* t .+ Œ∏‚ÇÄ)

Œ∏‚ÇÅ  = œÄ    # Phase shift in radians : radians to degrees -> Œ∏ * 180/œÄ

# Generate the cosine wave w‚ÇÅ
w‚ÇÅ = A * cos.(2œÄ * ùëì‚Çç .* t .+ Œ∏‚ÇÅ);
```

```julia
using Plots

# Signal parameters
A = 1      # Amplitude
f_c = 5    # Frequency, Hz
theta = œÄ / 4  # Phase shift, rad
t = range(0, stop=1, length=1000)  # Time vector

# Generate a cosine wave with a phase shift
signal = A * cos.(2 * œÄ * f_c * t .+ theta)

# Calculate amplitude and phase of the signal
amplitude = abs.(signal)
phase = angle.(signal)

# Create a plot with two subplots for amplitude and phase
p = plot(layout = (2, 1), size = (600, 400))

# Amplitude plot
plot!(p[1], t, amplitude, title = "Amplitude of the Signal", xlabel = "Time [s]", ylabel = "Amplitude")

# Phase plot
plot!(p[2], t, phase, title = "Phase of the Signal", xlabel = "Time [s]", ylabel = "Phase [rad]", ylims = (-œÄ, œÄ))

# Display the plot
display(p)
```
