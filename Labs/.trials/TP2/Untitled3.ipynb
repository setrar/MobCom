```julia
function lte_rs_gold(Ncp, Nid_cell)
    lte_gold_table = zeros(UInt32, 20, 2, 14)

    if Ncp == 0
        Nsymb = 4
        Ncp2 = 1
    else
        Nsymb = 3
        Ncp2 = 0
    end

    Nid_cellx2 = Nid_cell * 2

    for ns = 0:19
        for l = 0:1
            x2 = Ncp2 + Nid_cellx2 + ((1 + Nid_cellx2) * (1 + (Nsymb * l) + (7 * (1 + ns)))) * 1024
            # println("cinit (ns $ns, l $l) => $x2")
            x2_3 = x2 >>> 3
            x2_2 = x2 >>> 2
            x2_1 = x2 >>> 1
            x1 = 1 + (1 << 31)
            x2 = x2 ⊻ (x2 ⊻ (x2 ⊻ x2_3 ⊻ x2_2 ⊻ x2_1)) << 31
            for n = 1:49
                x1 = x1 ⊻ (x1 >>> 1) ⊻ (x1 >>> 4)
                x1 = x1 ⊻ (x1 ⊻ (x1 << 31)) ⊻ (x1 << 28)
                x2_4 = x2 >>> 4
                x2_3 = x2 >>> 3
                x2_2 = x2 >>> 2
                x2_1 = x2 >>> 1
                x2n = x2_1 ⊻ (x2_2 ⊻ (x2_3 ⊻ x2_4))
                x2 = x2n ⊻ (x2n ⊻ (x2n << 31)) ⊻ (x2n << 30) ⊻ (x2n << 29) ⊻ (x2n << 28)
            end

            for n = 0:13
                x1 = x1 ⊻ (x1 >>> 1) ⊻ (x1 >>> 4)
                x1 = x1 ⊻ (x1 ⊻ (x1 << 31)) ⊻ (x1 << 28)
                x2 = x2 ⊻ (x2 >>> 1) ⊻ (x2 >>> 2) ⊻ (x2 >>> 3) ⊻ (x2 >>> 4)
                x2 = x2 ⊻ (x2 ⊻ (x2 << 31)) ⊻ (x2 << 30) ⊻ (x2 << 29) ⊻ (x2 << 28)
                lte_gold_table[1 + ns, 1 + l, 1 + n] = x1 ⊻ x2
                # println("lte_gold_table[$(1 + ns)][$(1 + l)][$(1 + n)] = $(lte_gold_table[1 + ns, 1 + l, 1 + n])")
            end
        end
    end

    return lte_gold_table
end
```
