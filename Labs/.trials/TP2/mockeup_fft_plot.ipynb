```julia
using Plots
```

```julia
𝑓 = [abs(x) <= 1 ? 1 : 0 for x in -5:0.1:5];
```

```julia
plot(𝑓)
```

```julia
using FFTW
```

```julia
𝔽 = fft(𝑓); @show typeof(𝔽), length(𝔽);
# The fftshift() function conveniently rearranges the data 
# in the frequency domain so that the negative frequencies are on the left.
𝔽 = fftshift(𝔽);
N = length(𝑓);
Idx = LinRange(-0.5, 0.5, N+1)[1:N];  # normalized frequency 
```

```julia
plot(Idx, abs.(𝔽)
    # ,xlims=[-5,5]
    # ,ylims=[0,25]
    , xlabel="f[Hz]"
    , ylabel="Amplitude"
)
```

```julia
plot(Idx,10 * log10.(abs.(𝔽))
    # ,xlims=[-5,5]
    # ,ylims=[0,25]
    , xlabel="f[Hz]"
    , ylabel="Power [dB]"
)
```
