```julia
using Plots
```

```julia
ùëì = [abs(x) <= 1 ? 1 : 0 for x in -5:0.1:5];
```

```julia
plot(ùëì)
```

```julia
using FFTW
```

```julia
ùîΩ = fft(ùëì); @show typeof(ùîΩ), length(ùîΩ);
# The fftshift() function conveniently rearranges the data 
# in the frequency domain so that the negative frequencies are on the left.
ùîΩ = fftshift(ùîΩ);
N = length(ùëì);
Idx = LinRange(-0.5, 0.5, N+1)[1:N];  # normalized frequency 
```

```julia
plot(Idx, abs.(ùîΩ)
    # ,xlims=[-5,5]
    # ,ylims=[0,25]
    , xlabel="f[Hz]"
    , ylabel="Amplitude"
)
```

```julia
plot(Idx,10 * log10.(abs.(ùîΩ))
    # ,xlims=[-5,5]
    # ,ylims=[0,25]
    , xlabel="f[Hz]"
    , ylabel="Power [dB]"
)
```
