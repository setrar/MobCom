```julia
using SigMF
using Dates
```

```julia
data = zeros(ComplexF64, 1000);
```

```julia
data[100] = 1.0e2 + 0.0im
data[200] = 1.4e2 + 0.4im
data[300] = 1.8e2 + 0.8im
data[400] = 1.16e2 + 0.16im
data[500] = 1.32e2 + 0.23im
```

```julia
using Plots
```

```julia
plot(data)
```

```julia
# Create the metadata
sGlobal = SigMFGlobal(
    datatype = "cf32_le",
    sample_rate = 1.0e3,
    author = "me@you.com",
    description = "RX Signal with channel and frequency offset",
    version = "0.2.0" # SigMF.__version__,
);
```

```julia
# Create a capture key at time index 0
sCaptures = SigMFCaptures(
     frequency = 1.0e3
    ,datetime = Dates.format(Dates.now(), "yyyy-mm-dd HH:MM:SS")
);
```

```julia
meta = SigMFMeta(sGlobal,[sCaptures],[SigMFAnnotations()]);
```

```julia
file = SigMFFile(meta,data);
```

```julia
# Save metadata to disk
SigMF.save("../../data/white_random_noise.sigmf",file)
```
