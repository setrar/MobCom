Check the plotting functionality of Matplotlib. Here we show a simple static plot.

```julia
using Plots

t = -5:1/100:5
plot(t, sin.(t.^2) ./ (t.^2)
    , xlabel="\$t\$"
    , ylabel="\$\\sin(t^2)/(t^2)\$"
    , grid=true
)
```

```julia
using Plots

function showSin(f)
    t = -1:1/1000:1
    plot(t, sin.(2π*f*t.*t))
    plot!(t, cos.(2π*f./(t.^2 .+ 1)))
    plot!([0], [0.5], label="f=$f", legend=:topleft, framestyle=:box)
end

showSin(5)
display(Plots.current())
```
