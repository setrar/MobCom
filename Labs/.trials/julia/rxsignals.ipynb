````julia
using Plots

function plot_fir_signal(h)
    n = 0:length(h)-1
    scatter(n, h, xlabel="n", ylabel="h[n]", title="Finite-Impulse Response (FIR) Signal")
end

# Example FIR filter coefficients
fir_coefficients = [0.5, -0.2, 0.8, -0.4, 0.1]

# Plot the FIR signal
plot_fir_signal(fir_coefficients)
````
