If you want to translate the MATLAB code containing complex numbers into equivalent Python code, you should replace j with 1j in Python to represent the imaginary unit. Here's the equivalent Python code for the MATLAB snippet you provided:

```python
import numpy as np

# Define complex numbers in Python
pss_0 = np.array([-363 + 1j*0,
                  362 + 1j*0,
                  362 + 1j*0])

# Access the real and imaginary parts separately
real_parts = np.real(pss_0)
imaginary_parts = np.imag(pss_0)

# Print the results
for real, imag in zip(real_parts, imaginary_parts):
    print(f"{real} + {imag}j")
```

In this Python code:

- `1j` is used to represent the imaginary unit.
- `np.array` is used to create a NumPy array containing complex numbers.
- `np.real` and `np.imag` are used to access the real and imaginary parts of the complex numbers, respectively.
