````julia
using Plots

𝑓ₛ = 1000  # sampling_frequency
𝑓₍  = 5   # carrier_frequency

# Time vector
t = range(0, stop=1, length=𝑓ₛ)

# Carrier signal
𝑆₍ = sin.(2 * π * 𝑓₍ * t) # Carrier

# Data signal
data_frequency = 1
data = sin.(2 * π * data_frequency * t)

# AM Modulation
am = (1 .+ data) .* 𝑆₍

# FM Modulation
fm = sin.(2 * π * 𝑓₍ * t .+ data)

# PM Modulation
pm = sin.(2 * π * 𝑓₍ * t .+ π .* data)

# Plotting

plot(
      plot(t, am, title="Amplitude Modulation (AM)")
    , plot(t, fm, title="Frequency Modulation (FM)")
    , plot(t, pm, title="Phase Modulation (PM)")
    , layout = ( 3, 1)
    , xlabel="Time", ylabel="Amplitude", legend=false
    , ylims = (-2, 2)
    # , yticks = (-1, 1)
)
````

- Is channel modulation done in frequency domain?

Yes, channel modulation is indeed performed in the frequency domain. In signal processing, modulation involves manipulating a carrier signal's properties — such as its amplitude, frequency, or phase — in accordance with the information signal (the signal to be transmitted). This process inherently takes place in the frequency domain, as it is the characteristics of the carrier wave's frequency that are being altered.

To understand this better, consider these key points:

1. **Carrier Signal**: The carrier is a high-frequency signal that is used to carry the information. It is chosen such that it can efficiently propagate through the transmission medium (like air for radio waves, or optical fiber for fiber-optic communication).

2. **Information Signal**: This is the signal that contains the information to be transmitted. It is usually a low-frequency signal compared to the carrier.

3. **Modulation Process**: During modulation, certain properties of the carrier signal (amplitude, frequency, or phase) are varied in line with the information signal. This process effectively "encodes" the information onto the carrier wave.

4. **Frequency Domain**: The concept of altering a signal's amplitude, frequency, or phase is inherently a frequency domain concept. These properties are fundamental to how a signal is represented in the frequency domain.

5. **Purpose of Modulation**: One of the primary reasons for modulation is to allow the signal to be transmitted over a medium efficiently and to enable multiplexing (transmitting multiple signals over the same channel).

In summary, channel modulation is a process that takes place in the frequency domain, as it involves changing the characteristics of a carrier signal's frequency to encode and transmit information.

````julia
# Time vector
t = range(0, stop=1, length=1000)

# Carrier signals for QAM
carrier_frequency = 10
carrier1 = cos.(2 * π * carrier_frequency * t)
carrier2 = sin.(2 * π * carrier_frequency * t)

# Data signals for QAM
data_frequency1 = 1
data_frequency2 = 1
data1 = cos.(2 * π * data_frequency1 * t)
data2 = sin.(2 * π * data_frequency2 * t)

# QAM Modulation
qam = data1 .* carrier1 + data2 .* carrier2

# Plotting
plot(t, qam, title="Quadrature Amplitude Modulation (QAM)", xlabel="Time", ylabel="Amplitude", legend=false)
````
