```julia
using Plots

# Generate some sample data
t = 0:0.1:10
y1 = sin.(t)
y2 = cos.(t)
y3 = tan.(t)

# Create subplots
plot1 = plot(t, y1, label="sin(t)", xlabel="t", ylabel="y1")
plot2 = plot(t, y2, label="cos(t)", xlabel="t", ylabel="y2")
plot3 = plot(t, y3, label="tan(t)", xlabel="t", ylabel="y3")

# Display subplots
plot(plot1, plot2, plot3, layout=(3, 1), legend=true)
```

```julia
using Plots

# Generate some sample data
t = 0:0.1:10
y1 = sin.(t)
y2 = cos.(t)
y3 = tan.(t)

# Initialize an empty array to store the plots
plots = Plots.Plot[]

# Create plots and add them to the array using a loop
for (idx, series) in enumerate([y1, y2, y3])
    push!(plots, plot(t, series, label="Series $idx", xlabel="t", ylabel="y$idx"))
end

# Display the plots with a specific layout
plot(plots..., layout=(3, 1), legend=true)
```

```julia
using Plots

# Generate some sample data
t = 0:0.1:10
y1 = sin.(t)
y2 = cos.(t)
y3 = tan.(t)

# Initialize an empty array to store the plots
plots = Plots.Plot[]

# Create plots and add them to the array using a loop
for (idx, series) in enumerate([y1, y2, y3])
    p = plot(t, series, label="Series $idx", xlabel="t", ylabel="y$idx")
    annotate!(p, 5, 0.8, text("Index: $idx", :right, 10, :black))
    push!(plots, p)
end

# Display the plots with a specific layout
plot(plots..., layout=(3, 1), legend=true)
```

```julia
using Plots

# Generate some sample data
t = 0:0.1:10
y1 = sin.(t)
y2 = cos.(t)
y3 = tan.(t)

# Initialize an empty array to store the plots
plots = Plots.Plot[]

# Create plots and add them to the array using a loop
for (idx, series) in enumerate([y1, y2, y3])
    p = plot(t, series, label="Series $idx", xlabel="t", ylabel="y$idx")
    
    # Add title to the first subplot
    if idx == 1
        title!(p, "PSS Time Domain Correlation")
    end

    if idx == 3
        xlabel!(p, "Time (ms)")
    end

    annotate!(p, 5, 0.8, text("Index: $idx", :right, 10, :black))
    push!(plots, p)
end

# Display the plots with a specific layout
plot(plots..., layout=(3, 1), legend=true)
```
