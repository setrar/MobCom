```julia
function c_seq(init::Int, lengthy::Int)
    crop = 1600
    x1 = zeros(UInt8, crop + lengthy)
    x2 = zeros(UInt8, crop + lengthy)
    x1[1] = 1  # Julia arrays are 1-indexed

    for j in 0:30
        x2[j + 1] = (init >> j) & 0x01
    end

    for j in 1:(length(x1) - 31)
        x1[j + 31] = x1[j + 3] ⊻ x1[j]
        x2[j + 31] = x2[j + 3] ⊻ x2[j + 2] ⊻ x2[j + 1] ⊻ x2[j]
    end

    return x1[crop + 1:end] .⊻ x2[crop + 1:end]  # Apply XOR to the sliced arrays
end
```

```julia
init = 2113
sequence_lengthy = 288
sequence = c_seq(init, sequence_lengthy)

println(sequence)
```

```julia
function constellation_plot(z::Vector{ComplexF64}; title=nothing, ref="bpsk", ax=nothing, refscale=1, kwargs...)
    @show ref;
    ax = ax isa Nothing ? plot(size=(300, 300), legend=:false) : ax
    scatter!(ax, real(z), imag(z), kwargs...)
    plot_ref(ref, ax=ax, refscale=refscale)
    plot!(ax
        , aspect_ratio=:equal
        , xlims=(-1.5, 1.5), ylims=(-1.5, 1.5)
        , xticks=[], yticks=[]
        , framestyle=:box
    )
    # title && title!(ax, title)
    if title !== nothing
        title!(ax, title)
    end
end
```
