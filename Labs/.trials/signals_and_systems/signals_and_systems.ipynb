```julia

# Parameters for the approximation
t = range(0, stop=4, length=1000)  # Time vector
Œº = 2  # Center of the peak (where the delta function is located)
œÉ = 0.01  # Width of the Gaussian (make it very small)

# Gaussian function approximating the Dirac delta function
Œ¥ = exp.(-((t .- Œº).^2) / (2 * œÉ^2))

# Normalize to ensure the area under the curve is 1 (like the Dirac delta)
Œ¥ /= ‚àë(Œ¥) * (t[2] - t[1])

# Plot
p1 = plot(t, Œ¥, label=L"\delta(t-2)"
    , xlabel="t", ylabel="Amplitude", title="Approximation of Dirac Delta Function", linewidth=2
);
```

```julia
# Parameters
A = 1  # Amplitude
ùëì‚Çç = 5  # Frequency (carrier) in Hz
œï = 0  # Phase shift in radians
T = 1  # Total time in seconds
# t = 0:0.001:T  # Time vector from 0 to T seconds with a step of 0.001
```

```julia
# Generate the signal
x(t) = A * sin.(2œÄ * ùëì‚Çç * t .+ œï)

# Plot the signal
p2 = plot(t, x(t), label="Simple Sine Wave"
    , xlabel="Time (s)", ylabel="Amplitude"
    , title="Simple Sinusoidal Signal"
);
```

```julia
# Perform the convolution
y‚Çú = conv(Œ¥, x(t))  # "same" returns the central part of the convolution, same size as inputs

# Normalize the result for plotting
y‚Çú /= maximum(y‚Çú);
```

```julia
p3 = plot(y‚Çú
    , label="Convolution Result", title="Convolution of Signal 1 and 2"
);
```
