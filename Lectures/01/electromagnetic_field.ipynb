# Electromagnetic Field

In the **far field** (or **Fraunhofer region**) of an electromagnetic (EM) source, the electric field ($ E $) and magnetic field ($ H $) have specific characteristics and relationships that differ from those in the **near field** (or **Fresnel region**). Here's a summary of the differences between the electric and magnetic fields in the far field:

### 1. **Relationship Between $ E $ and $ H $**

   - In the far field, the electric and magnetic fields are **perpendicular to each other** and to the direction of propagation, forming a **transverse electromagnetic (TEM) wave**.
   - The ratio of the magnitudes of the electric field $ E $ to the magnetic field $ H $ is constant and is equal to the **intrinsic impedance of free space**, $ \eta_0 \approx 377 \, \Omega $:
     $
     \frac{E}{H} = \eta_0 \approx 377 \, \Omega
     $
   - This implies that, unlike in the near field, the electric and magnetic fields are tightly coupled and propagate together at the speed of light.

### 2. **Phase Relationship**

   - In the far field, the electric and magnetic fields are **in phase**. This means that they reach their maximum and minimum values simultaneously at any given point in space.
   - In the near field, the fields may not be in phase due to the complex interactions and reactive components, especially close to the antenna or source.

### 3. **Field Strength Dependence on Distance**

   - In the far field, both $ E $ and $ H $ fields decrease with distance as $ \frac{1}{r} $, where $ r $ is the distance from the source. This is due to the energy spreading over a spherical surface.
   - In the near field, the decay rates are more complex and vary for the electric and magnetic fields. For instance, some components may decay as $ \frac{1}{r^2} $ or $ \frac{1}{r^3} $, especially in the reactive near field.

### 4. **Radiative vs. Reactive Components**

   - In the far field, only **radiative components** of the electric and magnetic fields exist. This means the energy is carried away from the source and does not return, resulting in a net power flow that travels away from the source.
   - In the near field, there are significant **reactive components**. These do not radiate energy outward but instead store energy near the source, creating complex field patterns.

### 5. **Propagation Direction and Polarization**

   - The electric and magnetic fields in the far field are oriented perpendicular to the direction of propagation (forming a TEM wave).
   - The **polarization** of the wave (the direction of the electric field vector) remains constant as the wave propagates in the far field.

### Summary Table

| Property                        | Near Field                    | Far Field                      |
|---------------------------------|-------------------------------|--------------------------------|
| Relationship $ E/H $          | Complex and varying           | $ \frac{E}{H} = \eta_0 $     |
| Phase Relationship              | May not be in phase           | In phase                       |
| Distance Dependence             | $ \frac{1}{r^2} $ or $ \frac{1}{r^3} $ decay | $ \frac{1}{r} $ decay        |
| Radiative vs. Reactive          | Reactive components present   | Only radiative                 |
| Propagation                     | Complex                       | TEM (transverse electromagnetic) wave |

In essence, in the far field, the electric and magnetic fields behave in a predictable, coupled manner as a TEM wave, with a constant ratio and phase relationship, and decay at a rate of $ \frac{1}{r} $. In the near field, the relationship is more complex, with varying decay rates and significant reactive components.

In the **far field**:

- $ E $ and $ H $ fields are **perpendicular** to each other and to the direction of propagation (forming a **TEM wave**).
- $ E/H = \eta_0 \approx 377 \, \Omega $ (constant ratio, **intrinsic impedance of free space**).
- Fields are **in phase** and decay as $ \frac{1}{r} $.
- Only **radiative components** exist, with energy propagating away from the source.

In the **near field**:

- $ E $ and $ H $ fields have a **complex relationship** and may not be perpendicular or in phase.
- Fields can decay as $ \frac{1}{r^2} $ or $ \frac{1}{r^3} $.
- **Reactive components** dominate, storing energy near the source rather than radiating.

```julia
using Plots
using LaTeXStrings
```

```julia
# Define parameters
r  = 0.1:0.1:10  # Distance range in meters
𝘌₀ = 1.0        # Reference electric field strength (at r = 1 meter) \isansE
η₀ = 377      # Intrinsic impedance of free space in ohms \eta

# Far field decay (1/r)
𝘌_far = 𝘌₀ ./ r            # Electric field decay in the far field
𝘏_far = 𝘌_far ./ η₀      # Magnetic field in the far field (using E/H = eta0)

# Near field decay (1/r^2 and 1/r^3)
𝘌_near₁ = 𝘌₀ ./ r.^2       # Electric field decay for a reactive near field (1/r^2)
𝘌_near₂ = 𝘌₀ ./ r.^3       # Electric field decay for a reactive near field (1/r^3)

# $E \propto 1/r$ In

# Plot
p1 = plot(r, 𝘌_far, label=L"Far Field $(E \propto \frac{1}{r})$", xlabel="Distance (r)", ylabel="Field Strength"
    , lw=2, size = (700,300), legend= :none, title = "Electric and Magnetic Field Decay in Near and Far Field")
plot!(r, 𝘏_far, label=L"Far Field $(H \propto \frac{1}{r})$", lw=2, linestyle=:dash)
plot!(r, 𝘌_near₁, label=L"Near Field $(E \propto \frac{1}{r^2})$", lw=2, linestyle=:dot)
plot!(r, 𝘌_near₂, label=L"Near Field $(E \propto \frac{1}{r^3})$", lw=2, linestyle=:dashdot)


# Create a deep copy of p1 to p2
p2 = deepcopy(p1)

# Modify p2 without affecting p1
plot!(p2, ylim=(0, 120)
    , legend= :true, title = "(zoom)"
)

# Define a custom layout with relative heights
layout = @layout [grid(2, 1, heights=[0.6, 0.4])]

plot(p1, p2, layout=layout, size = (700,500))
```

# References

The notation $ E \propto \frac{1}{r} $ means that the **electric field strength $ E $** is **inversely proportional to the distance $ r $** from the source. This implies:

1. **Field Strength Decreases with Distance**: As the distance $ r $ from the source increases, the electric field $ E $ decreases. Specifically, if you double the distance $ r $, the electric field strength will be reduced by half.

2. **Spherical Wave Propagation**: This $ \frac{1}{r} $ relationship is characteristic of **spherical wave propagation** in the **far field**. When a wave radiates outward from a point source, it spreads over a spherical surface, and the intensity of the field decreases as the surface area of the sphere (which grows as $ 4 \pi r^2 $) increases.

3. **Energy Spreading**: Since the total energy radiated by the source is spread over a larger and larger area as $ r $ increases, the energy (and hence the field strength) per unit area decreases proportionally to $ \frac{1}{r^2} $. However, the electric field $ E $, which is related to the square root of power per unit area, decreases as $ \frac{1}{r} $.

In summary, $ E \propto \frac{1}{r} $ indicates that the electric field strength weakens as you move farther from the source, due to the spreading of the wave over a larger area in the far field.

#### does that H also represent the h in impulse reponse?

No, the \( H \) in electromagnetics (representing **magnetic field strength**) is different from the \( h \) in **impulse response** in signal processing.

- **\( H \) (magnetic field strength)**: In electromagnetics, \( H \) denotes the magnetic field strength and is measured in amperes per meter (A/m).

- **\( h(t) \) (impulse response)**: In signal processing, \( h(t) \) represents the **impulse response** of a system. It characterizes how the system responds over time to a brief input (an impulse) and is a function of time \( t \).

### Key Difference
- **\( H \)** in electromagnetics is related to the behavior of magnetic fields.
- **\( h(t) \)** in signal processing describes the time-domain response of a system to an impulse.

While they share a similar letter, they refer to completely different concepts in different fields.
