# Superposition (addition)

---
Here's an example of adding two sinusoids with different phases using Julia. We'll create two sinusoidal signals with the same frequency but different phases, add them together, and plot the result.

```julia
using Plots

# Define parameters
  = 1.0          # Frequency in Hz
 = 2 *     # Angular frequency in rad/s
t  = 0:0.01:2     # Time vector from 0 to 2 seconds with a 0.01 s step

# Define two sinusoids with different phases
A = 1.0                     # Amplitude of the first sinusoid
A = 1.0                     # Amplitude of the second sinusoid
 = 0.0                     # Phase of the first sinusoid in radians \phi
 =  / 3               # Phase of the second sinusoid in radians

# Create the sinusoidal signals
y(t) = A * sin.( * t .+ )
y(t) = A * sin.( * t .+ )

# Add the two sinusoids
y(t) = y.(t) .+ y.(t)

# Plot the results
plot(t, y.(t), label="Sinusoid 1 (Phase = 0)", lw=2)
plot!(t, y.(t), label="Sinusoid 2 (Phase = /3)", lw=2, linestyle=:dash)
plot!(t, y.(t), label="Sum of Sinusoids"
    , xlabel = "Time (s)", ylabel = "Amplitude"
    , title = "Superposition of Two Sinusoids with Different Phases"
    , lw=2, linestyle=:dot
)
```

### Explanation

- **$ f $**: Frequency of the sinusoids (1 Hz).
- **$ _0 $**: Angular frequency ($ 2f $).
- **$ A_1 $, $ A_2 $**: Amplitudes of the two sinusoids.
- **$\phi_1$** and **$\phi_2$**: Phases of the two sinusoids, set to 0 and $ \frac{\pi}{3} $ respectively.

This code will plot three curves:

1. **Sinusoid 1**: The first sinusoidal signal with a phase of 0.
2. **Sinusoid 2**: The second sinusoidal signal with a phase of $ \frac{\pi}{3} $.
3. **Sum of Sinusoids**: The result of adding the two sinusoids, which shows the combined waveform due to superposition.

The resulting plot will demonstrate how the different phases affect the sum, with constructive or destructive interference at different points in time.
