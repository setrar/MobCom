# Sinc Function

```julia
using Plots
```

```julia
# Define the sinc function
function sinc(x)
    return x == 0 ? 1.0 : sin(π * x) / (π * x)
end
```

```julia
# Generate values
x = -10:0.01:10  # Range of x values
y = @. sinc(x)  # Vectorized operation

# Plot the sinc function
plot(x, y
    , label="sinc(x)", title="Normalized Sinc Function"
    , xlabel="x", ylabel="sinc(x)"
    , legend=:topright
)
```
