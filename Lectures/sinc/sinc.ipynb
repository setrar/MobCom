# Sinc Function

```julia
using Plots
```

```julia
# Define the normalized sinc function
function sinc(x)
    return x == 0 ? 1.0 : sin(π * x) / (π * x)
end
```

```julia
# Define the unormalized sinc function
function usinc(x)
    return x == 0 ? 1.0 : sin(x) / (x)
end
```

```julia
# Generate values
x = -20:0.01:20  # Range of x values
y = @. sinc(x)  # Vectorized operation
yᵤ = @. usinc(x)  # Vectorized operation

# Plot the sinc function
plot(x, y
    , label="sinc(x)", title="Sinc Functions"
    , xlabel="x", ylabel="sinc(x)"
    , legend=:topright
)
plot!(x, yᵤ
    , label="usinc(x)"
)
```
