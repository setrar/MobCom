```julia
using LinearAlgebra
using Plots

# Define two vectors x_A and x_B
x_A = [4.0, 3.0]  # Point/vector A
x_B = [1.0, 1.0]  # Point/vector B

# Compute the difference vector v = x_A - x_B
v = x_A - x_B

# Normalize the vector v to obtain the unit direction vector
v_unit = v / norm(v)

# Function to project a vector onto the direction of v_unit
function projection(v_unit, y)
    return (dot(v_unit, y) / dot(v_unit, v_unit)) * v_unit
end

# Example of a vector y to be projected onto the direction of v_unit
y = [3.0, 2.0]

# Compute the projection of y onto v_unit
proj_y_onto_v = projection(v_unit, y)

# Plot the vectors
plot(arrow = :closed, legend = :topleft)
plot!([0, x_A[1]], [0, x_A[2]], label = "x_A", arrow = true, lw = 2, color = :blue)
plot!([0, x_B[1]], [0, x_B[2]], label = "x_B", arrow = true, lw = 2, color = :green)
plot!([x_B[1], x_A[1]], [x_B[2], x_A[2]], label = "v = x_A - x_B", arrow = true, lw = 2, color = :red)
plot!([0, v_unit[1]], [0, v_unit[2]], label = "Unit vector v", arrow = true, lw = 2, linestyle = :dash, color = :purple)
plot!([0, y[1]], [0, y[2]], label = "y", arrow = true, lw = 2, color = :orange)
plot!([0, proj_y_onto_v[1]], [0, proj_y_onto_v[2]], label = "Projection of y onto v", arrow = true, lw = 2, color = :black)

# Set plot axis limits and labels
plot!(
    xlims!(-1, 5), ylims!(-1, 4)
    , xlabel!("x"), ylabel!("y")
    )

# Display the plot
# display(plot)
```
