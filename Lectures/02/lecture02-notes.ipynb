# References

The key difference between **Linear Time-Invariant (LTI)** systems and **Linear Time-Variant (LTV)** systems lies in their dependence on time:

### **Linear Time-Invariant (LTI) Systems**:
1. **Time-Invariance**: The system's behavior and characteristics do not change over time. If the input to the system is shifted in time, the output will shift by the same amount, but otherwise remain unchanged.
   - Mathematically: If $ x(t) \to y(t) $, then $ x(t - \tau) \to y(t - \tau) $.
2. **Impulse Response**: The response of the system to an impulse input ($ \delta(t) $) is fixed and does not vary with time.
3. **Simpler Analysis**: LTI systems can be analyzed using convolution, Fourier transform, Laplace transform, or Z-transform.
4. **Frequency Domain**: They have a well-defined frequency response that does not vary with time.

   **Example**:
   - An electrical circuit with constant resistors, capacitors, and inductors.

---

### **Linear Time-Variant (LTV) Systems**:
1. **Time-Variance**: The system's characteristics (e.g., coefficients, behavior) change with time. The system's response to the same input signal may be different depending on when the input is applied.
   - Mathematically: If $ x(t) \to y(t) $, then $ x(t - \tau) $ does not necessarily lead to $ y(t - \tau) $.
2. **Impulse Response**: The impulse response varies with time and is denoted as $ h(t, \tau) $, where $ t $ is the current time and $ \tau $ is the time of the impulse.
3. **More Complex Analysis**: LTV systems require time-varying methods, like state-space representation or specific time-domain approaches.
4. **Frequency Domain**: They do not have a fixed frequency response.

   **Example**:
   - A system where parameters change over time, such as a communication channel with time-varying fading or an RLC circuit with a time-dependent resistor.

---

### **Comparison Summary**:

| **Property**          | **LTI Systems**                      | **LTV Systems**                      |
|-----------------------|--------------------------------------|--------------------------------------|
| Time Dependency       | Does not change with time           | Changes with time                   |
| Impulse Response      | Fixed ($ h(t - \tau) $)            | Time-varying ($ h(t, \tau) $)     |
| Frequency Response    | Well-defined                        | Not well-defined                    |
| Analysis Complexity   | Simpler (Fourier, Laplace, Z-transform) | More complex (state-space, time-varying methods) |
| Example               | Constant RLC circuit                | Time-varying fading communication channel |
